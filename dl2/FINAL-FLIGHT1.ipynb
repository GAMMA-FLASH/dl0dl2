```python
nd = 36
xm = 18000
spectra_energy_L1=pL1.spectra_energy(ndivbin=nd, xmax=xm)
spectra_energy_G1=pG1.spectra_energy(ndivbin=nd, xmax=xm)
spectra_energy_H1=pH1.spectra_energy(ndivbin=nd, xmax=xm)
spectra_energy_I1=pI1.spectra_energy(ndivbin=nd, xmax=xm)
spectra_energy_J1=pJ1.spectra_energy(ndivbin=nd, xmax=xm)
spectra_energy_flight1 = pJ1._sum(spectra_energy_H1, spectra_energy_I1, normalize=True)
spectra_energy_flight1 = pJ1._sum(spectra_energy_flight1, spectra_energy_J1, normalize=True)

spectra_energy_L2=pL2.spectra_energy(ndivbin=nd, xmax=xm)
spectra_energy_G2=pG2.spectra_energy(ndivbin=nd, xmax=xm)
spectra_energy_H2=pH2.spectra_energy(ndivbin=nd, xmax=xm)
spectra_energy_I2=pI2.spectra_energy(ndivbin=nd, xmax=xm)
spectra_energy_J2=pJ2.spectra_energy(ndivbin=nd, xmax=xm)
spectra_energy_flight2 = pJ2._sum(spectra_energy_H2, spectra_energy_I2, normalize=True)
spectra_energy_flight2 = pJ2._sum(spectra_energy_flight2, spectra_energy_J2, normalize=True)

spectra_energy_L3=pL3.spectra_energy(ndivbin=nd, xmax=xm)
spectra_energy_G3=pG3.spectra_energy(ndivbin=nd, xmax=xm)
spectra_energy_H3=pH3.spectra_energy(ndivbin=nd, xmax=xm)
spectra_energy_I3=pI3.spectra_energy(ndivbin=nd, xmax=xm)
spectra_energy_J3=pJ3.spectra_energy(ndivbin=nd, xmax=xm)
spectra_energy_flight3 = pJ3._sum(spectra_energy_H3, spectra_energy_I3, normalize=True)
spectra_energy_flight3 = pJ3._sum(spectra_energy_flight3, spectra_energy_J3, normalize=True)

spectra_energy_flight = pJ2._sum(spectra_energy_flight1, spectra_energy_flight2, normalize=True)
spectra_energy_flight = pJ2._sum(spectra_energy_flight, spectra_energy_flight3, normalize=True)
spectra_energy_flightH = pJ2._sum(spectra_energy_H1, spectra_energy_H2, normalize=True)
spectra_energy_flightH = pJ2._sum(spectra_energy_flightH, spectra_energy_H3, normalize=True)
spectra_energy_flightI = pJ2._sum(spectra_energy_I1, spectra_energy_I2, normalize=True)
spectra_energy_flightI = pJ2._sum(spectra_energy_flightI, spectra_energy_I3, normalize=True)
spectra_energy_flightJ = pJ2._sum(spectra_energy_J1, spectra_energy_J2, normalize=True)
spectra_energy_flightJ = pJ2._sum(spectra_energy_flightJ, spectra_energy_J3, normalize=True)
spectra_energy_ground = pJ2._sum(spectra_energy_G1, spectra_energy_G2, normalize=True)
spectra_energy_ground = pJ2._sum(spectra_energy_ground, spectra_energy_G3, normalize=True)

c=pJ._drawH([spectra_energy_flightH, spectra_energy_flightI, spectra_energy_flightJ], nameCanvas="spectra_energy", title="Energy RPG1 RPG2 RPG3", xtitle="keV")
energies=[511, 609, 1120, 1460, 1764, 2614]
le=[]
for energy in energies:
    la = ROOT.TLine(energy, 0, energy, 1)
    la.SetLineStyle(7)
    la.SetLineColor(2)
    la.Draw()
    le.append(la)

c.Draw()
```
